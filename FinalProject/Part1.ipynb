Global radiation refers to the total solar radiation received on a horizontal surface, encompassing both direct and diffuse solar radiation. <br>
This radiation is a key component of Earth's energy balance and plays a crucial role in various natural processes, including weather patterns, climate dynamics, and ecosystem functioning. <br><br>
In the context of time-series analysis, global radiation data recorded over time forms a time series, where each data point represents the radiation level measured at a specific time interval (e.g., daily, hourly). <br>
Analyzing this time series allows for the identification of patterns, trends, and seasonal variations in solar radiation, providing insights into solar energy availability, climate variability, and long-term environmental changes.<br>
Moreover, understanding the temporal dynamics of global radiation is essential for optimizing renewable energy systems, such as solar power generation, and for studying the impacts of climate change on Earth's energy budget and ecosystems.

```python
import pandas as pd
import numpy as np
import statsmodels as sm
import seaborn as sns
import scipy as sc
import datetime as dt
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

import warnings
warnings.filterwarnings("ignore")
```

**Data Origin:**

https://ims.gov.il/he/node/272

```python
file_path = "data/global_rad_technion_2017_2023.csv"
data = pd.read_csv(file_path,parse_dates=['date'])
data['avg_daily_radiation'] = data.mean(axis='columns',numeric_only=True)
data['date_column'] = pd.to_datetime(data['date'],format='mixed',dayfirst=True) 
data['date_string'] = data['date_column'].dt.strftime('%d-%m-%Y')
data[['day','month','year']] = data['date_string'].str.split('-', expand=True)
qmap = {1:'Q1', 2:'Q2', 3:'Q3', 4:'Q4'}
data['quarter'] = data['date_column'].dt.quarter.map(qmap)
data.head()
```

```python
data.shape
```

```python
mini_data = data[['date_column','avg_daily_radiation','year','month','day','quarter']].set_index('date_column')
mini_data.head()
```

#### Seasonal Decomposition

```python
df = mini_data
palette = {'darknavy':'#001253','mid_blue':'#3E6D9C','mid_orange':'#FD841F','light_blue':'#7FC7D9','dark_orange':'#E14D2A',}
palette2 = {'green_teel':'#24A19C','lilach':"#9D65C9",'peach':'#FE7E6D','banana':"#FFB961"}
palette3 = ['#FF9800',"#2C7865","#124076","#FFD0EC"]
order = sorted(df['quarter'].unique())
```

```python
import matplotlib.dates as mdates

plt.figure(figsize=(20, 6))
plt.plot(df.index, df['avg_daily_radiation'], color='#0A1D56')
plt.title('Daily Average Global Radiation (01-01-2017 to 31-12-2023)')
plt.xlabel('Date')
plt.ylabel('Global Radiation\nkWh/m^2')

plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%y'))
plt.show()
```

```python
decomposition = seasonal_decompose(df['avg_daily_radiation'], period=365)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid



plt.figure(figsize=(8, 8))
plt.suptitle('Seasonal Decomposition of Global Radiation Time Series')
plt.subplot(411)

plt.plot(df.index, df['avg_daily_radiation'], label='Original', color='#001253') #darknavy
plt.legend(loc='best')
plt.subplot(412)
plt.title('Trend')
plt.plot(df.index, trend, label='Trend', color='#365486') #mid_blue
plt.legend(loc='best')
plt.subplot(413)
plt.title('Seasonality')
plt.plot(df.index, seasonal, label='Seasonal', color='#51829B') #dirt blue
plt.legend(loc='upper right')
plt.subplot(414)
plt.title("Residuals")
plt.plot(df.index, residual, label='Residual', color='#7FC7D9') #dark_orange
plt.legend(loc='upper right')

plt.tight_layout()
plt.savefig("decompose.png")
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(20, 10))
plot_acf(df['avg_daily_radiation'], lags=200)
plt.title('Autocorrection Plot')
plt.xlabel('Lag')
plt.ylabel('Autocorrection')
plt.tight_layout()
plt.show()
```

```python
palette2 = {'Q4':'#BB3B0E','Q3':"#0A81AB",'Q2':'#E45826','Q1':"#001253"}
mean_values = df.groupby('quarter')['avg_daily_radiation'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(x='avg_daily_radiation',data=df, hue='quarter', shade=True, palette=reversed(list(palette.values())),common_norm=False,hue_order=order,legend=True,linewidth=2,alpha=.6 )
pos = {'Q4':0.002,'Q3':0.004,'Q2':0.002,'Q1':0.0045}
plt.legend(title='Quarter',loc='upper right',labels=['Q1', 'Q2', 'Q3', 'Q4'])
for i, row in mean_values.iterrows():
    mean_value = row['avg_daily_radiation']
    quarter_color = palette2[row['quarter']]
    y = pos[row['quarter']]
    plt.axvline(x=mean_value, color=quarter_color, linestyle='--', linewidth=1.2,)
    plt.text(mean_value,y, f"{mean_value:.2f}", rotation=0, verticalalignment='bottom', horizontalalignment='center',color=quarter_color,bbox=dict(facecolor='white', alpha=1, edgecolor=quarter_color, boxstyle='round,pad=0.25'))

plt.xlabel('Global Radiation\nkWh/m^2')
plt.title("Global Radiation KDE per Quarter (Means)")
plt.savefig("KDE.png")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
#sns.boxplot(data=df, y='avg_daily_radiation', x='quarter', palette=palette2.values(), saturation=1)
sns.boxplot(y='avg_daily_radiation',data=df, x='quarter',hue='quarter', palette=list(palette.values()),hue_order=order,gap=.2,legend='full',linewidth=1.5,saturation=0.8,medianprops={"color": "black", "linewidth": 1 ,"linestyle":'--'})
plt.legend(title='Quarter',loc='upper right',)# ncol=4)
plt.ylabel('Global Radiation\nkWh/m^2')
plt.xlabel('Quarter')
#plt.tight_layout()
plt.title("Global Radiation Box-plot per Quarter")
plt.savefig("box.png")
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.catplot(data=df,x='avg_daily_radiation',y='quarter', hue="quarter", aspect=2.5,palette=list(palette2.values()),hue_order=order)

plt.ylabel('Global Radiation\nkWh/m^2')
plt.xlabel('Quarter')
plt.title("Global Radiation Box-plot per Quarter")
plt.legend(title='Quarter', loc='upper right', labels=['Q1', 'Q2', 'Q3', 'Q4'])
#plt.tight_layout()
plt.savefig("catplot.png")
plt.show()
```
